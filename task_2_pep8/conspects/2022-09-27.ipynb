```python
%%markdown
**Области видимости**

1. встроенные функции, типы данных и коллекции (`__builtins__`)
2. глобальная область видимости (все могут получить доступ к глобальным переменным, объектам и методам)
3. локальная область видимости (к этим переменным и объектам имеют доступ только переменные и объекты внутри тела функции или класса, в котором они определены)
4. вложенная область видимости (область видимости внешней функции относительно дочерних функций)
```

```python
def sample_local():
    a = 6
    print(a)
```

```python
sample_local()
```

```python
def sample_change_global():
    global a
    print(f'Old value {a=}')
    a = 7
    print(f'New value {a=}')
```

```python
sample_change_global()
```

```python
def outer_func():
    outer_var = 100 # это как раз переменная с вложенной видимостью
    print(f'Outer func: {outer_var=}')
    def inner_func():
        print(f'Inner func: {outer_var=}')
    inner_func()
    print(f'Outer func: {outer_var=}')
# вложенная видимость позволяет только читать переменные из внешней функции, но не изменять их 
```

```python
outer_func()
```

```python
def outer_func_nonlocal():
    outer_var = 100 # это как раз переменная с вложенной видимостью
    print(f'Outer func: {outer_var=}')
    def inner_func():
        nonlocal outer_var
        print(f'Inner func: {outer_var=}')
        outer_var = 200
    inner_func()
    print(f'Outer func: {outer_var=}')
```

```python
outer_func_nonlocal()
```
